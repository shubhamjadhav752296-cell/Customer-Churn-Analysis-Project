```python
sns.countplot(x='gender', hue='Churn', data=data)
plt.title("Gender-wise Customer Count")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
ax = sns.countplot(x='SeniorCitizen', data=data)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
 # Create cross-tab
ct = pd.crosstab(data['gender'], data['Churn'])

# Convert to percentage
ct_perc = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = ct_perc.plot(
    kind='bar',
    stacked=True,
    figsize=(6,4)
)

# Add percentage labels
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    if height > 0:
        ax.text(
            x + width/2,
            y + height/2,
            f'{height:.1f}%',
            ha='center',
            va='center',
            fontsize=9,
            color='white'
        )

plt.title("Gender-wise Churn Percentage")
plt.xlabel("Gender")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

plt.figure(figsize=(18, 12))

for i, col in enumerate(cols):
    plt.subplot(3, 3, i + 1)
    sns.countplot(x=col, data=data, hue = 'Churn')
    plt.title(col)
    plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

#Customers without add-on services like OnlineSecurity, TechSupport, DeviceProtection, and OnlineBackup show a significantly higher churn rate compared to those who have these services.
#Churn is also higher among users with fiber optic InternetService than DSL.
#Customers with No PhoneService or No MultipleLines contribute less overall but still show noticeable churn.
#Overall, having bundled services is strongly associated with lower customer churn.
